```python
df = df.head(200)
```

```python
import plotly_express as px
px.bar(df, x='alcaldia_hechos')
```

```python
centro = [19.399712350357127, -99.18985396940174]
import folium
map = folium.Map(location=centro, zoom_start=10, titles='cartodbpositron', attr='lol')
```

```python
map
```

```python
subset = df.head(50)
for delito,lat,lon in zip(subset['delito'], subset['latitud'], subset['longitud']):
    folium.Marker(location=[lat, lon], popup=delito, tooltip='puchale aqui').add_to(map)
```

```python
up = [19.373806429449715, -99.18579846265872]
folium.Circle(location=up, radius=250, fill_color='red').add_to(map)
folium.Marker(location=up, icon=folium.Icon(color='green', icon='info-sign'), popup='UP').add_to(map)
```
