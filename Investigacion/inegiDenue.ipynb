```python
#https://www.inegi.org.mx/app/api/denue/v1/consulta/Buscar/camiones/21.85717833,-102.28487238/250/[aquí va tu Token]
#5da04172-9e3d-43d9-99bf-e7c0a5e2d09f

def busqueda(termino, radio, latitud, longitud):
    import requests 
    import pandas as pd
    url = f'https://www.inegi.org.mx/app/api/denue/v1/consulta/Buscar/{termino}/{latitud},{longitud}/{radio}/5da04172-9e3d-43d9-99bf-e7c0a5e2d09f'
    resultado = requests.get(url)
    return pd.DataFrame(resultado.json())
```

```python
resultados = busqueda('tortas',5000,19.295727195723405,-99.21060812033221)
resultados
```

```python
import folium
six = [19.295727195723405,-99.21060812033221]
map = folium.Map(location=six, zoom_start=15)
```

```python
folium.Circle(location=six, radius=250, fill_color='blue').add_to(map)
folium.Marker(location=six, icon=folium.Icon(color='green', icon='info-sign'), popup='SIx Flags').add_to(map)
```

```python

subset = resultados.head(50)
for torta,lon,lat in zip(subset['Nombre'], subset['Longitud'], subset['Latitud']):
    folium.Marker(location=[lon, lat], popup=torta).add_to(map)
```
