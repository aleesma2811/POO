```python
#Alejandra Escallada

import requests
import pandas as pd
url_base =  "https://hp-api.onrender.com/api"
response = requests.get(url=f"{url_base}/characters")

class hpAPI():
    def __init__(self):
        print("Objeto creado")
        self.__url_base = "https://hp-api.onrender.com/api"

    def personajes(self):
        import requests
        response = requests.get(f"{self.__url_base}/characters")
        return pd.DataFrame(response.json())

    def hechizos(self):
        import requests
        response = requests.get(f"{self.__url_base}/spells")
        return pd.DataFrame(response.json())
    
    def personajes_por_casa(self):
        import requests
        response = requests.get(f"{self.__url_base}/characters")
        return pd.DataFrame(response.json())['house'].value_counts()

    def personajes_por_genero(self):
        import requests
        response = requests.get(f"{self.__url_base}/characters")
        return pd.DataFrame(response.json())['gender'].value_counts()
    
    def ancestry(self):
        import requests
        response = requests.get(f"{self.__url_base}/characters")
        return pd.DataFrame(response.json())['ancestry'].value_counts()
    
    def personajes_vivos_muertos(self, x):
        import requests
        response = requests.get(f"{self.__url_base}/characters")
        return pd.DataFrame(response.json())['alive'].apply(lambda x: 'Vivo' if x == True else 'Muerto')
    
    def filrar_por_casa(self, casa):
        import requests
        try:
            response = requests.get(f"{self.__url_base}/characters/house/{casa}")
            return pd.DataFrame(response.json())
        except:
            return "err"

    
print(response)
```
