```python
titulo = df[df['channel_title']=='Franco Escamilla'].sort_values(by='dislikes', ascending=False).iloc[0]['title'] #iloc: localizar por indice
dislikes = df[df['channel_title']=='Franco Escamilla'].sort_values(by='dislikes', ascending=False).iloc[0]['dislikes'] #iloc: localizar por indice
print(f"El video con más dislikes es {titulo} con {dislikes} dislikes")
```
